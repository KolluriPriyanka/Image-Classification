```python
Path = "/content/drive/MyDrive/Flowers Dataset"
```

```python
flowers = {}
```

```python
for dirname, dirlist, filenames in os.walk(Path):
    flower_name = dirname.split('/')[-1]
    if dirname != Path:
        print(f"Number of {flower_name} images: {len(filenames)}")
        filePaths = []
        for filename in filenames:
            filePaths.append(os.path.join(Path, dirname, filename))
        flowers[flower_name] = filePaths
```

```python
fig, ax = plt.subplots(1,3, figsize=(24, 12))

img1 = cv2.imread(flowers[list(flowers.keys())[0]][0])
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
ax[0].imshow(img1)

img2 = cv2.imread(flowers[list(flowers.keys())[0]][1])
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
ax[1].imshow(img2)

img3 = cv2.imread(flowers[list(flowers.keys())[0]][2])
img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
ax[2].imshow(img3)

plt.show()
```

```python
fig, ax = plt.subplots(1,3, figsize=(24, 12))

for idx, img in enumerate([img1, img2, img3]):

    img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

    ax[idx].imshow(img_gray, cmap='gray')

plt.show()
```

```python
fig, ax = plt.subplots(1,3, figsize=(24, 6))

for idx, img in enumerate([img1, img2, img3]):
    color = ('b','g','r')
    for i, col in enumerate(color):
        histr = cv2.calcHist([img], [i], None, [256], [0,256])
        ax[idx].plot(histr, color = col)
    ax[idx].set_title('Image 1')

plt.show()
```
